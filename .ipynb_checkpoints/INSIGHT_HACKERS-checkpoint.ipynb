## Step 5 - Preprocessing

```python
sc = StandardScaler()
df_scaled = sc.fit_transform(df.drop(columns=['diagnosis']))
```

```python
df_scaled = pd.DataFrame(df_scaled, columns=df.columns[1:])
```

```python
df_scaled.head()
```

```python
df_scaled.isnull().sum()
```

```python
df_scaled = df_scaled.drop(columns=['Unnamed: 32'])
```

```python
df_scaled.shape
```

#### BEFORE REMOVING OUTLIERS

#### K_MEANS

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, n_init=10, random_state=42)
  kmeans.fit(df_scaled)
  wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,11), wcss, marker='o', linestyle='-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, n_init=10, random_state=42)
df_scaled['CLUSTER'] = kmeans.fit_predict(df_scaled)
```

```python
df_scaled.head()
```

```python
from sklearn.decomposition import PCA

# Silhouette Score
sil_score = silhouette_score(df_scaled.drop(columns=['CLUSTER']), df_scaled['CLUSTER'])
print(f"\nSilhouette Score: {sil_score}")
```

## Step 6 - Model Selection

```python
y=df['diagnosis']
x=df.drop(['diagnosis','Unnamed: 32'],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

#### Model-Training

```python
LR = LogisticRegression(max_iter=1000)
LR.fit(x_train,y_train)
y_pred=LR.predict(x_test)
print(f"Accuracy Score:",accuracy_score(y_pred,y_test)*100)
print(f"Training Score:",LR.score(x_train,y_train)*100)
print(f"Confusion Matirx:\n",confusion_matrix(y_test,y_pred))

print(classification_report(y_test,y_pred))
```
