# üìä Dataset Description: Healthcare Data  

## üè• Overview  
This dataset contains **medical diagnostic data** used for analyzing **health conditions** (potentially cancer-related). The dataset includes **569 records** with **30 numerical features**, which describe various characteristics of cell nuclei in medical scans.  

## üìå Data Attributes  
Each row in the dataset represents a **patient's medical report**, with the following key attributes:  

### **1. Tumor Size & Shape Features**  
- `radius_mean` - Mean radius of the tumor.  
- `texture_mean` - Mean texture (variation in gray-scale values).  
- `perimeter_mean` - Mean perimeter of the tumor.  
- `area_mean` - Mean tumor area.  
- `smoothness_mean` - Mean smoothness (local variation in radius lengths).  

### **2. Compactness & Concavity Features**  
- `compactness_mean` - Mean compactness (perimeter¬≤/area - 1.0).  
- `concavity_mean` - Mean concavity (severity of concave portions).  
- `concave points_mean` - Mean number of concave portions of the contour.  

### **3. Symmetry & Fractal Dimension Features**  
- `symmetry_mean` - Mean symmetry of the tumor shape.  
- `fractal_dimension_mean` - Mean fractal dimension (coastline approximation - 1).  

### **4. Standard Error Features (Measurement Variability)**  
- `radius_se`, `texture_se`, `perimeter_se`, `area_se`, `smoothness_se`, etc.  
- These features represent the **standard error** of each measurement, indicating **variability** in tumor characteristics.  

### **5. Worst Case Scenario Features (Largest Tumor Cell Observations)**  
- `radius_worst`, `texture_worst`, `perimeter_worst`, `area_worst`, etc.  
- These capture the **worst (largest) values recorded** for each tumor measurement.  

## üìå Missing Data  
- The dataset **does not contain any missing values**, ensuring a **clean and structured dataset** for machine learning applications.

**Level 1: Data Preprocessing**

**Outlier Detection**

Identification of Outliers

Outliers are data points that significantly deviate from the rest of the dataset. Identifying outliers is crucial to ensure data quality and prevent skewed analysis.

Chosen Method: **Z-score**

We use the Z-score method to identify outliers. The Z-score measures how many standard deviations a data point is from the mean:



A common threshold for detecting outliers is:

Z > 3 or Z < -3 (data points beyond 3 standard deviations from the mean are considered outliers).

This method is chosen because our dataset follows a normal distribution, making Z-score an appropriate metric for identifying extreme values.

Approach for Handling Outliers

After detecting outliers, we take the following approach:

Capping: If an outlier is a valid but extreme observation, we cap it at a predefined threshold (e.g., ).

Removal: If the outlier is an error or an anomaly, we remove it from the dataset.

Impact on Data Quality

Reduces skewness: Eliminating extreme values helps maintain a balanced dataset.

Improves model performance: Machine learning models become more stable and less sensitive to extreme values.

Enhances interpretability: A cleaned dataset ensures meaningful insights without the influence of extreme values.


```python
#compute Z-scores = x - u(mean) / standard deviation(sd)
z_scores = (df[numerical_cols] - df[numerical_cols].mean()) / df[numerical_cols].std()
outliers_zscore = (np.abs(z_scores) > 3).sum()
```

**WE HAVE USED Z-SCORE NORMALIZATION BECAUSE GIVEN DATSET IS NORMAL AS SHOWN IN THE HISTPLOT**

```python
outlier_report = pd.DataFrame({ 'Z_score Outliers': outliers_zscore})
print(outlier_report)
```

```python
from scipy.stats import zscore

# Compute Z-scores for numerical columns
z_scores = df[numerical_cols].apply(zscore)

# Define threshold (commonly 3)
threshold = 3

# Filter the data
df_cleaned = df[(z_scores.abs() < threshold).all(axis=1)]
```

```python
print(f"Original dataset size: {df.shape}")
print(f"Dataset size after z-score outlier removal: {df_cleaned.shape}")
```

**Data normalization**

there are 2 scaling techniques : Z-score and Min-max scaling
Z-score standarization = X - mean / standard dev
Min-max scaling = X-Xmin / Xmax -Xmin

**Since our dataset is normal so we'll use z-score norm**

```python
df_standardized = df_cleaned.copy()

for col in numerical_cols:
    mean = df[col].mean()
    std = df[col].std()
    df_standardized[col] = (df[col] - mean) / std

print(df_standardized.head())  # Check transformed data
```

**Data Preprocessing done!!**

Level 2: Data Analysis & Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


# Initialize k-means parameters
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "random_state": 1,
}

# Create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df_standardized)
    sse.append(kmeans.inertia_)

# Visualize results
plt.plot(range(1, 11), sse, marker='o')
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE (Sum of Squared Errors)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

**As we can see in the graph elbow point is 2**

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 11):  # Start from 2 since silhouette isn't defined for k=1
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    labels = kmeans.fit_predict(df_standardized)
    score = silhouette_score(df_standardized, labels)
    silhouette_scores.append(score)

# Plot Silhouette Score for different k values
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle="--")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score Analysis")
plt.show()
```

**using the silhoutte score we can see the quality of clusters**

```python
print("Original DataFrame shape:", df.shape)
print("Standardized DataFrame shape:", df_standardized.shape)
print("Missing values in df:\n", df.isnull().sum())
```

**K-mean Clustering**

```python
# Fit KMeans with optimal clusters
optimal_k = 2
kmeans = KMeans(n_clusters=optimal_k, **kmeans_kwargs)
df_standardized['Cluster'] = kmeans.fit_predict(df_standardized)

# Print cluster centers
print("Cluster Centers (in standardized form):\n", kmeans.cluster_centers_)
```

```python
# Compute Silhouette Score
silhouette_avg = silhouette_score(df_standardized, df_standardized['Cluster'])
print(f"Silhouette Score for k={optimal_k}: {silhouette_avg:.4f}")
```

```python

from sklearn.decomposition import PCA

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)  # Reduce to 2 principal components for visualization
pca_result = pca.fit_transform(df_standardized.drop('Cluster', axis=1))

# Create a new DataFrame with PCA results
df_pca = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
df_pca['Cluster'] = df_standardized['Cluster']

# Visualize PCA results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=df_pca, palette='viridis')
plt.title('PCA Visualization of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Explained Variance Ratio
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained Variance Ratio: {explained_variance_ratio}")
print(f"Total Explained Variance: {sum(explained_variance_ratio)}")
```
