Here we conclude that the data is not gaussian. And There are Many Unwanted Anomalies(Outliers) are present.
To get rid of those anamolies we will use Turkey's IQR method.

```python
# method for detection and removal of anamolies
def IQR_method (df,n,features):
    """
    Takes a dataframe and returns an index list corresponding to the observations 
    containing more than n outliers according to the Tukey IQR method.
    """
    outlier_list = []
    
    for column in features:
                
        # 1st quartile (25%)
        Q1 = np.percentile(df[column], 25)
        # 3rd quartile (75%)
        Q3 = np.percentile(df[column],75)
        
        # Interquartile range (IQR)
        IQR = Q3 - Q1
        
        # outlier step
        outlier_step = 1.5 * IQR
        
        # Determining a list of indices of outliers
        outlier_list_column = df[(df[column] < Q1 - outlier_step) | (df[column] > Q3 + outlier_step )].index
        
        # appending the list of outliers 
        outlier_list.extend(outlier_list_column)
        
    # selecting observations containing more than x outliers
    outlier_list = Counter(outlier_list)        
    multiple_outliers = list( k for k, v in outlier_list.items() if v > n )
    
    # Calculate the number of records below and above lower and above bound value respectively
    df1 = df[df[column] < Q1 - outlier_step]
    df2 = df[df[column] > Q3 + outlier_step]
    
    print('Total number of outliers is:', df1.shape[0]+df2.shape[0])
    
    return multiple_outliers
```

```python
Outliers_IQR = IQR_method(df_raw, 1, feature_list)
df_out = df_raw.drop(Outliers_IQR,axis=0).reset_index(drop=True)
```

Lets also do the Z_score method to see which gives us the best results

```python
def ZScore_method(df, n, features):
    """
    Takes a dataframe df of features and returns an index list corresponding to the observations 
    containing more than n outliers according to the z-score method.
    """
    outlier_list = []
    
    for column in features:
        # calculate the mean and standard deviation of the data frame
        data_mean = df[column].mean()
        data_std = df[column].std()
        threshold = 3
        
        z_score = abs( (df[column] - data_mean)/data_std )
        
        # Determining a list of indices of outliers for feature column        
        outlier_list_column =  df[z_score > threshold].index
        
        # appending the found outlier indices for column to the list of outlier indices 
        outlier_list.extend(outlier_list_column)
        
    # selecting observations containing more than x outliers
    outlier_list = Counter(outlier_list)        
    multiple_outliers = list( k for k, v in outlier_list.items() if v > n )
    
    # Calculate the number of outlier records
    df1 = df[z_score > threshold]
    print('Total number of outliers is:', df1.shape[0])
    
    return multiple_outliers
```

```python
z_score_outliers = ZScore_method(df_raw,1, feature_list)
df_out1 = df_raw.drop(z_score_outliers, axis =0).reset_index(drop=True)
```

```python
# plotting the graps to see if we got rid of outliers or not
fig, axes = plt.subplots(nrows=3, ncols=3,figsize=(11,17))
fig.suptitle('Features vs diagnosis\n', size = 18)

sns.boxplot(ax=axes[0,0], data=df_out1, x='diagnosis', y='texture_se')
axes[0,0].set_title("texture_se distribution")

sns.boxplot(ax=axes[0, 1], data=df_out1, x='diagnosis', y='radius_se')
axes[0,1].set_title("radius_se distribution")

sns.boxplot(ax=axes[0, 2], data=df_out1, x='diagnosis', y='area_worst')
axes[0,2].set_title("area_worst distribution")

sns.boxplot(ax=axes[1, 0], data=df_out1, x='diagnosis', y='concavity_worst')
axes[1,0].set_title("concavity_worst distribution")

sns.boxplot(ax=axes[1, 1], data=df_out1, x='diagnosis', y='concave points_worst')
axes[1,1].set_title("concave points_worst distribution")

sns.boxplot(ax=axes[1, 2], data=df_out1, x='diagnosis', y='compactness_se')
axes[1,2].set_title("compactness_se distribution")

sns.boxplot(ax=axes[2, 0], data=df_out1, x='diagnosis', y='smoothness_mean')
axes[2,0].set_title("smoothness_mean distribution")

sns.boxplot(ax=axes[2, 1], data=df_out1, x='diagnosis', y='fractal_dimension_se')
axes[2,1].set_title("fractal_dimension_se distribution")

sns.boxplot(ax=axes[2, 2], data=df_out1, x='diagnosis', y='area_se')
axes[2,2].set_title("area_se distribution")

plt.tight_layout()
```

```python
X = df_out.drop('diagnosis', axis=1)
y = df_out['diagnosis']
```

```python
# implementing Kmeans on the data to cluster them.
# before implementing kmeans we have nomalize the data since we are using kmeans its better to use minmax normalisation technique
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
X_scaled = ms.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.head()
```

***Using the elbow method to find out best no of cluster***

```python
from sklearn.cluster import KMeans

# elbow method
cs = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    cs.append(kmeans.inertia_)
plt.plot(range(1,11),cs)
plt.title('Elbow Method \n')
plt.xlabel('No of clusters \n')
plt.ylabel('CS')
plt.show()
```

Here there is a kink in no of cluster 2 so we try implementing n_cluster = 2

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X_scaled)

centroids = kmeans.cluster_centers_
lebels = kmeans.labels_

plt.scatter(X_scaled_df['radius_mean'], X_scaled_df['texture_mean'],c=lebels, s=20, cmap='viridis')
plt.scatter(centroids[:,0], centroids[:,1], c='red',s=100,marker="o",label="Centroids")
plt.legend()
plt.show()
```

Since the dataset is not 2d i.e n-dimentional , we can implement pca to reduce the dimentions to 2 then implement k means for better result

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled_df)

clusters = kmeans.fit_predict(X_pca)
plt.scatter(X_pca[:,0],X_pca[:,1],c=clusters,s=20,alpha=0.6,cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='red',marker='o',label='Centroid')
plt.legend()
plt.show()
```

After preprocessing the data is ready to be trained for training we can use random forest classification technique


```python
# importing the libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=42)
# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=3, criterion='entropy', max_depth=None, random_state=42)

# Train the model
rf_classifier.fit(x_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100)

# Optional: Detailed classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
